```python
import matplotlib.pyplot as plt
import json
js=open('qafsechuman.json', 'r')
dictionary = json.load(js)

for key,value in dictionary.items():
    print(key)
    print(value)
    ## LINE GRAPH ##
    x=value['subplots']['replicahumansec000/fepsechuman.out']['xdata']
    y=value['subplots']['replicahumansec000/fepsechuman.out']['ydata']
    fig=plt.figure()

    plt.plot(x,y, color='blue', marker='o')
    plt.xlabel('E1-E2  [kcal/mol]')
    plt.ylabel('Free energy  [kcal/mol]')
    plt.title('Sec Human')
    plt.text(0.05, 0.95, 'Alpha shift   Hij ',
         horizontalalignment='center',
         verticalalignment='top',
         transform=plt.gca().transAxes,
         fontsize=7)
    plt.show()
    break

```
